```python
#Importing Libraries
import keras 
import pandas as pd 
import numpy as np 
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
#Reading Data as a csv file
Concrete_Data=pd.read_csv("concrete_data.csv")
```

```python
# Exploring Data
Concrete_Data.head()
```

```python
#Exploring Shape of Data
Concrete_Data.shape
```

```python
# Exploring Information of Concrete_Data
Concrete_Data.describe()
```

```python
#Exploring if There is Null Values or Not
Concrete_Data.isnull().sum()
```

```python
# Splitting X Values and Y Values
Concrete_Data_Columns = Concrete_Data.columns
Prediction = Concrete_Data[Concrete_Data_Columns[Concrete_Data_Columns != 'Strength']]
Target = Concrete_Data['Strength']
```

```python
# Exploring X Values
Prediction.head()
```

```python
#Exploring Y Values
Target.head()
```

```python
#Normalizing Dataset
PredictionNorm=(Prediction-Prediction.mean())/Prediction.std()
PredictionNorm.head()
```

```python
# Creating Regression Model using Keras
def Regression_Model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(PredictionNorm.shape[1],)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
# Looping over Epochs
# Produced Lesser Mean than B
beginning=0
end=50
for i in range(beginning , end):
  Train_Data,Test_Data,Y_Train,Y_Test = train_test_split(PredictionNorm,Target,test_size=0.3,random_state=i)
  Reg_Model=Regression_Model()
  Reg_Model.fit(Train_Data,Y_Train,epochs=100,verbose=2)
  Y_Pred=Reg_Model.predict(Test_Data)
  MSE=[]
  MSE.append(metrics.mean_squared_error(Y_Test,Y_Pred))
  i=i+1
print("The Mean of MSEs is : "+str(np.mean(MSE)))
print("The Standard of MSEs is : "+str(np.std(MSE)))
```
